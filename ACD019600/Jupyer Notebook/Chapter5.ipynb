## 重複值處理

```python
df = pd.read_excel(r"C:\ACD019600\data\test5.xlsx", sheet_name="5-2")
df
```

```python
df.drop_duplicates()
```

```python
df
```

```python
df.drop_duplicates(subset = "唯一識別碼")
```

```python
df.drop_duplicates(subset = ["客戶姓名","唯一識別碼"])
```

```python
#保留最後一個重複值
df.drop_duplicates(subset = ["客戶姓名","唯一識別碼"], keep = "last")
```

```python
#不保留任何重複值
df.drop_duplicates(subset = ["客戶姓名","唯一識別碼"], keep = False)
```

## 索引設定
### 為無索引表添加索引

```python
df = pd.read_excel("C:/ACD019600/data/test5.xlsx", sheet_name="5-5")
df
```

```python
# 為表添加欄索引
df.columns = ["訂單編號","客戶姓名","唯一識別碼","成交時間"]
df
```

```python
#為表添加列索引
df.index = [1,2,3,4,5]
df
```

```python
df.set_index("訂單編號")
```

### 重新命名索引

```python
df = pd.read_excel("C:/ACD019600/data/test5.xlsx", sheet_name="5-4")
df.index = [1,2,3,4,5,6]
df
```

```python
#重新命名欄索引
df.rename(columns = {"訂單編號":"新訂單編號", "客戶姓名":"新客戶姓名"})
```

```python
#重新命名列索引
df.rename(index = {1:"一",
                   2:"二",
                   3:"三"})
```

```python
#同時重新命名列索引和欄索引
df.rename(columns = {"訂單編號":"新訂單編號",
                     "客戶姓名":"新客戶姓名"},
            index = {1:"一",
                     2:"二",
                     3:"三"})
```
